```python
"""
Target: filter out the not right triple in GenWiki
"""

import ast
from tqdm import tqdm

train_list = []
train_tar_list = []
test_list = []
test_tar_list = []
with open('./test.source', 'r') as f, open('./test.target', 'r') as f2:
    lines = f.readlines()
    lines2 = f2.readlines()
    i = 0
    for line in tqdm(lines):
        try:
            test_list.append(ast.literal_eval(line.strip()))
            test_tar_list.append(lines2[i].strip())
            i += 1
        except:
            i += 1
            continue

with open('./train.source', 'r') as f, open('./train.target', 'r') as f2:
    lines = f.readlines()
    lines2 = f2.readlines()
    i = 0
    for line in tqdm(lines):
        try:
            train_list.append(ast.literal_eval(line.strip()))
            train_tar_list.append(lines2[i].strip())
            i += 1
        except:
            i += 1
            continue

with open('./test_new.source', 'w') as f:
    for line in tqdm(test_list):
        f.write(str(line) + '\n')
with open('./train_new.source', 'w') as f:
    for line in tqdm(train_list):
        f.write(str(line) + '\n')
with open('./test_new.target', 'w') as f:
    for line in tqdm(test_tar_list):
        f.write(str(line) + '\n')
with open('./train_new.target', 'w') as f:
    for line in tqdm(train_tar_list):
        f.write(str(line) + '\n')
```
